```python
# multidimensional arrays
import numpy as np
import math
import pyarma as pa

# inline plots
import matplotlib.pyplot as plt
%matplotlib inline

# nicer figures
import matplotlib as mpl
mpl.rcParams['axes.titlesize'] = 18
mpl.rcParams['axes.labelsize'] = 16
mpl.rcParams['xtick.labelsize'] = 14
mpl.rcParams['ytick.labelsize'] = 14
mpl.rcParams['legend.fontsize'] = 16
plt.rcParams["figure.figsize"] = (7.5,5)
```

```python
# Reading data from binary files without a double-slit barrier
prob = np.loadtxt("probability.txt", skiprows=2, usecols=0, dtype='double')
t = np.loadtxt("probability.txt", skiprows=2, usecols=1, dtype='double')

# Reading data from binary files with a double-slit barrier
prob_ds = np.loadtxt("probability_ds.txt", skiprows=2, usecols=0, dtype='double')
```

```python
# Compute the absolute error of the probability without a double-slit barrier
abs_err = abs(prob - np.ones(np.size(prob)))

# Compute the absolute error of the probability with a double-slit barrier
abs_err_ds = abs(prob_ds - np.ones(np.size(prob)))
```

```python
# Plot of the deviation of probability as a function of time
plt.plot(t, abs_err, label="without double-slit barrier")
plt.plot(t, abs_err_ds, label="with double-slit barrier")

# add axis labels and legend
plt.xlabel("$t$")
plt.ylabel("Absolute error")
plt.legend()

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)

# save plot as pdf
plt.savefig("abs_err.pdf")
```
