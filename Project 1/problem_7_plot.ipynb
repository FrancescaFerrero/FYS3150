```python
# multidimensional arrays
import numpy as np
import math

# inline plots
import matplotlib.pyplot as plt
%matplotlib inline

# nicer figures
import matplotlib as mpl
mpl.rcParams['axes.titlesize'] = 16
mpl.rcParams['axes.labelsize'] = 14
mpl.rcParams['legend.fontsize'] = 12
mpl.rcParams['xtick.labelsize'] = 12
mpl.rcParams['ytick.labelsize'] = 12
#mpl.rcParams['axes.spines.right'] = False
#mpl.rcParams['axes.spines.top'] = False
```

```python
#reading file with n=10
file = open('solutions_7_10pts.txt', 'r')
lines = file.readlines() 
file.close()

x_num_10=[]
v_10=[]

for line in lines:
    x_num_10.append(line.split()[0])
    v_10.append(line.split()[1])

for i in range(len(x_num_10)):
    x_num_10[i]=float(x_num_10[i])
    
for i in range(len(v_10)):
    v_10[i]=float(v_10[i])

#reading file with n=100

file = open('solutions_7_100pts.txt', 'r')
lines = file.readlines() 
file.close()

x_num_100=[]
v_100=[]

for line in lines:
    x_num_100.append(line.split()[0])
    v_100.append(line.split()[1])

for i in range(len(x_num_100)):
    x_num_100[i]=float(x_num_100[i])
    
for i in range(len(v_100)):
    v_100[i]=float(v_100[i])
    
#reading file with n=1000

file = open('solutions_7_1000pts.txt', 'r')
lines = file.readlines() 
file.close()

x_num_1000=[]
v_1000=[]

for line in lines:
    x_num_1000.append(line.split()[0])
    v_1000.append(line.split()[1])

for i in range(len(x_num_1000)):
    x_num_1000[i]=float(x_num_1000[i])
    
for i in range(len(v_1000)):
    v_1000[i]=float(v_1000[i])
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(1, 3, figsize=(16, 5))

# plot theoretical result
x_cont = np.linspace(start=0, stop=1, num=10000)
u=[1-(1-math.exp(-10))*x_cont[i]-math.exp(-10*x_cont[i]) for i in range(10000)]
ax1.plot(x_cont,u,color="red", label="Exact $\it{u(x)}$",linewidth=0.6)
ax2.plot(x_cont,u,color="red", label="Exact $\it{u(x)}$",linewidth=0.6)
ax3.plot(x_cont,u,color="red", label="Exact $\it{u(x)}$",linewidth=0.6)

# plot numerical result
ax1.scatter(x_num_10,v_10,color="blue", label="Numerical $\it{v(x)}$",s=2)
ax2.scatter(x_num_100,v_100,color="blue", label="Numerical $\it{v(x)}$",s=1.)
ax3.scatter(x_num_1000,v_1000,color="blue", label="Numerical $\it{v(x)}$",s=0.2)

# add axis labels and legend
ax1.set_xlabel("x")
ax1.set_ylabel("Solution")
ax1.legend()
ax2.set_xlabel("x")
ax2.set_ylabel("Solution")
ax2.legend()
ax3.set_xlabel("x")
ax3.set_ylabel("Solution")
ax3.legend()
# add a title 
ax1.set_title("n=10")
ax2.set_title("n=100")
ax3.set_title("n=1000")

ax1.grid(linestyle = '--', linewidth = 0.1)
ax2.grid(linestyle = '--', linewidth = 0.1)
ax3.grid(linestyle = '--', linewidth = 0.1)

plt.savefig("fig_7b.pdf")
```
