```python
# multidimensional arrays
import numpy as np
import math

# inline plots
import matplotlib.pyplot as plt
%matplotlib inline

# histogram
import seaborn as sbn

# nicer figures
import matplotlib as mpl
mpl.rcParams['axes.titlesize'] = 18
mpl.rcParams['axes.labelsize'] = 16
mpl.rcParams['xtick.labelsize'] = 14
mpl.rcParams['ytick.labelsize'] = 14
mpl.rcParams['legend.fontsize'] = 16
plt.rcParams["figure.figsize"] = (7,5)
```

```python
#random initial state___ for histogram

# T=1
energy_1 = np.loadtxt("6_rand_1_20.txt", usecols=0, dtype='double')

# T=2.4
energy_2 = np.loadtxt("6_rand_2_20.txt", usecols=0, dtype='double')
```

```python
# Plotting normalized histogram - random initial state - T=1
sbn.histplot(energy_1, bins=30, stat="density", color="blue")

# add axis labels and legend
plt.xlabel("$\epsilon$ [J]")
plt.ylabel("$p_{\epsilon}$($\epsilon;T$) [1/J]")
#plt.legend()

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)
plt.title("$L=20$ $T=1$ [J/$k_B$]")

# save plot as pdf
plt.savefig("histogram_1.pdf")

# Computing and printing variance
print(np.var(energy_1))
print(np.average(energy_1))
```

```python
# Plotting normalized histogram - random initial state - T=2.4
sbn.histplot(energy_2, bins=40, stat="density", color="red")

# add axis labels and legend
plt.xlabel("$\epsilon$ [J]")
plt.ylabel("$p_{\epsilon}$($\epsilon;T$) [1/J]")
#plt.legend()

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)
plt.title("$L=20$ $T=2.4$ [J/$k_B$]")

# save plot as pdf
plt.savefig("histogram_2.pdf")

# Computing and printing variance
print(np.var(energy_2))
print(np.average(energy_2))
```
