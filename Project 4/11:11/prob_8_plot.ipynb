```python
# multidimensional arrays
import numpy as np
import math

# inline plots
import matplotlib.pyplot as plt
%matplotlib inline

# nicer figures
import matplotlib as mpl
mpl.rcParams['axes.titlesize'] = 18
mpl.rcParams['axes.labelsize'] = 16
mpl.rcParams['xtick.labelsize'] = 14
mpl.rcParams['ytick.labelsize'] = 14
mpl.rcParams['legend.fontsize'] = 16
plt.rcParams["figure.figsize"] = (7,5)
```

```python
# ordered initial state
#reading temperatures, average energy and magnetization, specific heat capacity and susceptibility

# L=20
T = np.loadtxt("8_ord_20.txt", usecols=0, dtype='double')
energy_20 = np.loadtxt("8_ord_20.txt", usecols=1, dtype='double')
magn_20 = np.loadtxt("8_ord_20.txt", usecols=2, dtype='double')
Cv_20 = np.loadtxt("8_ord_20.txt", usecols=3, dtype='double')
X_20 = np.loadtxt("8_ord_20.txt", usecols=4, dtype='double')

# L=40
energy_40 = np.loadtxt("8_ord_40.txt", usecols=1, dtype='double')
magn_40 = np.loadtxt("8_ord_40.txt", usecols=2, dtype='double')
Cv_40 = np.loadtxt("8_ord_40.txt", usecols=3, dtype='double')
X_40 = np.loadtxt("8_ord_40.txt", usecols=4, dtype='double')

# L=60
energy_60 = np.loadtxt("8_ord_60.txt", usecols=1, dtype='double')
magn_60 = np.loadtxt("8_ord_60.txt", usecols=2, dtype='double')
Cv_60 = np.loadtxt("8_ord_60.txt", usecols=3, dtype='double')
X_60 = np.loadtxt("8_ord_60.txt", usecols=4, dtype='double')

# L=80
energy_80 = np.loadtxt("8_ord_80.txt", usecols=1, dtype='double')
magn_80 = np.loadtxt("8_ord_80.txt", usecols=2, dtype='double')
Cv_80 = np.loadtxt("8_ord_80.txt", usecols=3, dtype='double')
X_80 = np.loadtxt("8_ord_80.txt", usecols=4, dtype='double')

# L=100
energy_100 = np.loadtxt("8_ord_100.txt", usecols=1, dtype='double')
magn_100 = np.loadtxt("8_ord_100.txt", usecols=2, dtype='double')
Cv_100 = np.loadtxt("8_ord_100.txt", usecols=3, dtype='double')
X_100 = np.loadtxt("8_ord_100.txt", usecols=4, dtype='double')
```

```python
# Plotting energies 
plt.plot(T, energy_20, color="blue",label="L=20")
plt.plot(T, energy_40, color="red",label="L=40")
plt.plot(T, energy_60, color="green",label="L=60")
plt.plot(T, energy_80, color="yellow",label="L=80")
plt.plot(T, energy_100, color="magenta",label="L=100")

# add axis labels and legend
plt.xlabel("T [J/k$_B$]")
plt.ylabel("<$\epsilon$> [J]")
plt.legend()

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)

# save plot as pdf
plt.savefig("Energy.pdf")
```

```python
# Plotting magnetization 
plt.plot(T, magn_20, color="blue",label="L=20")
plt.plot(T, magn_40, color="red",label="L=40")
plt.plot(T, magn_60, color="green",label="L=60")
plt.plot(T, magn_80, color="yellow",label="L=80")
plt.plot(T, magn_100, color="magenta",label="L=100")

# add axis labels and legend
plt.xlabel("T [J/k$_B$]")
plt.ylabel("<|m|>")
plt.legend()

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)

# save plot as pdf
plt.savefig("Magnetization.pdf")
```

```python
# Plotting specific heat capacity 
plt.plot(T, Cv_20, color="blue",label="L=20")
plt.plot(T, Cv_40, color="red",label="L=40")
plt.plot(T, Cv_60, color="green",label="L=60")
plt.plot(T, Cv_80, color="yellow",label="L=80")
plt.plot(T, Cv_100, color="magenta",label="L=100")

# add axis labels and legend
plt.xlabel("T [J/k$_B$]")
plt.ylabel("C$_v$ [k$_B$/J]")
plt.legend()

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)

# save plot as pdf
plt.savefig("Heat_capacity.pdf")
```

```python
# Plotting susceptibility
plt.plot(T, X_20, color="blue",label="L=20")
plt.plot(T, X_40, color="red",label="L=40")
plt.plot(T, X_60, color="green",label="L=60")
plt.plot(T, X_80, color="yellow",label="L=80")
plt.plot(T, X_100, color="magenta",label="L=100")

# add axis labels and legend
plt.xlabel("T [J/k$_B$]")
plt.ylabel("$\chi$ [1/J]")
plt.legend()

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)

# save plot as pdf
plt.savefig("Susceptibility.pdf")
```
