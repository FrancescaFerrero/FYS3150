```python
# multidimensional arrays
import numpy as np
import math

# inline plots
import matplotlib.pyplot as plt
%matplotlib inline

# histogram
import seaborn as sbn

# nicer figures
import matplotlib as mpl
mpl.rcParams['axes.titlesize'] = 18
mpl.rcParams['axes.labelsize'] = 16
mpl.rcParams['xtick.labelsize'] = 14
mpl.rcParams['ytick.labelsize'] = 14
mpl.rcParams['legend.fontsize'] = 16
plt.rcParams["figure.figsize"] = (7,5)
```

```python
#random initial state
#reading number of cycles, average energy and magnetization per particle L=2

# T=1
n_cycles = np.loadtxt("5a_rand_1_2.txt", usecols=0, dtype='double')
energy_1 = np.loadtxt("5a_rand_1_2.txt", usecols=1, dtype='double')
magn_1 = np.loadtxt("5a_rand_1_2.txt", usecols=2, dtype='double')

# T=2.4
energy_2 = np.loadtxt("5a_rand_2_2.txt", usecols=1, dtype='double')
magn_2 = np.loadtxt("5a_rand_2_2.txt", usecols=2, dtype='double')
```

```python
#random initial state
#reading number of cycles, average energy and magnetization per particle L=20

# T=1
n_cycles_20 = np.loadtxt("5a_rand_1_20.txt", usecols=0, dtype='double')
energy_1_20 = np.loadtxt("5a_rand_1_20.txt", usecols=1, dtype='double')
magn_1_20 = np.loadtxt("5a_rand_1_20.txt", usecols=2, dtype='double')

# T=2.4
energy_2_20 = np.loadtxt("5a_rand_2_20.txt", usecols=1, dtype='double')
magn_2_20 = np.loadtxt("5a_rand_2_20.txt", usecols=2, dtype='double')
```

```python
# Plotting energies - random initial state L=2
plt.plot(n_cycles, energy_1, color="blue",label="T=1 [J/$k_b$]")
plt.plot(n_cycles, energy_2, color="red",label="T=2.4 [J/$k_b$]")

# add axis labels and legend
plt.xlabel("Number of MC cycles")
plt.ylabel("<$\epsilon$> [J]")
plt.legend()

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)
plt.title("Random initial state L=2")

# save plot as pdf
plt.savefig("Energy_rand_2.pdf")
```

```python
# Plotting energies - random initial state L=20
plt.plot(n_cycles_20, energy_1_20, color="blue",label="T=1 [J/$k_b$]")
plt.plot(n_cycles_20, energy_2_20, color="red",label="T=2.4 [J/$k_b$]")

# add axis labels and legend
plt.xlabel("Number of MC cycles")
plt.ylabel("<$\epsilon$> [J]")
plt.legend()

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)
plt.title("Random initial state L=20")

# save plot as pdf
plt.savefig("Energy_rand_20.pdf")
```

```python
# Plotting magnetiziation - random initial state L=2
plt.plot(n_cycles, magn_1, color="blue",label="T=1 [J/$k_b$]")
plt.plot(n_cycles, magn_2, color="red",label="T=2.4 [J/$k_b$]")

# add axis labels and legend
plt.xlabel("Number of MC cycles")
plt.ylabel("<|m|>")
plt.legend()

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)
plt.title("Random initial state L=2")

# save plot as pdf
plt.savefig("magn_rand_2.pdf")
```

```python
# Plotting magnetiziation - random initial state L=20
plt.plot(n_cycles_20, magn_1_20, color="blue",label="T=1 [J/$k_b$]")
plt.plot(n_cycles_20, magn_2_20, color="red",label="T=2.4 [J/$k_b$]")

# add axis labels and legend
plt.xlabel("Number of MC cycles")
plt.ylabel("<|m|>")
plt.legend()

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)
plt.title("Random initial state L=20")

# save plot as pdf
plt.savefig("magn_rand_20.pdf")
```

```python
# Plotting normalized histogram - random initial state
sbn.histplot(energy_1_20, bins=32, stat="density", color="blue",label="T=1 [J/$k_b$]")
sbn.histplot(energy_2_20, bins=32, stat="density", color="red",label="T=2.4 [J/$k_b$]")

# add axis labels and legend
plt.xlabel("<$\epsilon$> [J]")
plt.ylabel("p$_{\epsilon}$($\epsilon$;T)")
plt.legend()

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)
plt.title("Random initial state L=20")

# save plot as pdf
plt.savefig("histogram_rand.pdf")

# Computing and printing variances
print(np.var(energy_1_20))
print(np.var(energy_2_20))
```
