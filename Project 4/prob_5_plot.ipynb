```python
# multidimensional arrays
import numpy as np
import math

# inline plots
import matplotlib.pyplot as plt
%matplotlib inline

# nicer figures
import matplotlib as mpl
mpl.rcParams['axes.titlesize'] = 18
mpl.rcParams['axes.labelsize'] = 16
mpl.rcParams['xtick.labelsize'] = 14
mpl.rcParams['ytick.labelsize'] = 14
mpl.rcParams['legend.fontsize'] = 16
plt.rcParams["figure.figsize"] = (7,5)
```

```python
#reading number of cycles, average energy and magnetization per particle

# T=1
n_cycles = np.loadtxt("5a_rand_1.txt", usecols=0, dtype='double')
energy_1 = np.loadtxt("5a_rand_1.txt", usecols=1, dtype='double')
magn_1 = np.loadtxt("5a_rand_1.txt", usecols=2, dtype='double')

# T=2.4
energy_2 = np.loadtxt("5a_rand_2.txt", usecols=1, dtype='double')
magn_2 = np.loadtxt("5a_rand_2.txt", usecols=2, dtype='double')
```

```python
# Plotting energies _ random initial state
plt.plot(n_cycles, energy_1, color="blue",label="T=1 [J/$k_b$]")
plt.plot(n_cycles, energy_2, color="red",label="T=2.4 [J/$k_b$]")

# add axis labels and legend
plt.xlabel("Number of MC cycles")
plt.ylabel("Energy [J]")
plt.legend()

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)
plt.title("Random initial state")

# save plot as pdf
plt.savefig("Energy_random.pdf")
```

```python
# Plotting magnetiziation _ random initial state
plt.plot(n_cycles, magn_1, color="blue",label="T=1 [J/$k_b$]")
plt.plot(n_cycles, magn_2, color="red",label="T=2.4 [J/$k_b$]")

# add axis labels and legend
plt.xlabel("Number of MC cycles")
plt.ylabel("Magnetization")
plt.legend()

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)
plt.title("Random initial state")

# save plot as pdf
plt.savefig("magn_random.pdf")
```
