```python
# multidimensional arrays
import numpy as np
import math

# inline plots
import matplotlib.pyplot as plt
%matplotlib inline

# histogram
import seaborn as sbn

# nicer figures
import matplotlib as mpl
mpl.rcParams['axes.titlesize'] = 18
mpl.rcParams['axes.labelsize'] = 16
mpl.rcParams['xtick.labelsize'] = 14
mpl.rcParams['ytick.labelsize'] = 14
mpl.rcParams['legend.fontsize'] = 16
plt.rcParams["figure.figsize"] = (7,5)
```

```python
#random initial state
#reading number of cycles, average energy and magnetization per spin L=20

# T=1
n_cycles_20 = np.loadtxt("5a_rand_1_20.txt", usecols=0, dtype='double')
energy_1_20 = np.loadtxt("5a_rand_1_20.txt", usecols=1, dtype='double')
magn_1_20 = np.loadtxt("5a_rand_1_20.txt", usecols=2, dtype='double')

# T=2.4
energy_2_20 = np.loadtxt("5a_rand_2_20.txt", usecols=1, dtype='double')
magn_2_20 = np.loadtxt("5a_rand_2_20.txt", usecols=2, dtype='double')
```

```python
### ordered initial state
#reading number of cycles, average energy and magnetization per particle L=20

# T=1
n_cycles_20_ord = np.loadtxt("5a_ord_1_20.txt", usecols=0, dtype='double')
energy_1_20_ord = np.loadtxt("5a_ord_1_20.txt", usecols=1, dtype='double')
magn_1_20_ord = np.loadtxt("5a_ord_1_20.txt", usecols=2, dtype='double')

# T=2.4
energy_2_20_ord = np.loadtxt("5a_ord_2_20.txt", usecols=1, dtype='double')
magn_2_20_ord = np.loadtxt("5a_ord_2_20.txt", usecols=2, dtype='double')
```

```python
# Plotting energies - T=2.4 L=20

plt.plot(n_cycles_20_ord, energy_2_20_ord, color="blue",label="Initially ordered")
plt.plot(n_cycles_20, energy_2_20, color="red",label="Initially random")

# add axis labels and legend
plt.xlabel("Number of MC cycles")
plt.ylabel(r"$\langle\epsilon\rangle$ [J]")
plt.legend(loc='upper right')

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)
plt.title("T=2.4 [J/$k_b$] L=20")
plt.xscale("log")

# save plot as pdf
plt.savefig("Energy_2_20.pdf")
```

```python
# Plotting energies - T=1 L=20
plt.plot(n_cycles_20_ord, energy_1_20_ord, color="blue",label="Initially ordered")
plt.plot(n_cycles_20, energy_1_20, color="red",label="Initially random")

#plt.plot(n_cycles_20_ord, energy_2_20_ord, color="red",label="T=2.4 [J/$k_b$]")

# add axis labels and legend
plt.xlabel("Number of MC cycles")
plt.ylabel("<$\epsilon$> [J]")
plt.legend(loc='upper right')

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)
plt.title("T=1 [J/$k_b$] L=20")
plt.xscale("log")

# save plot as pdf
plt.savefig("Energy_1_20.pdf")
```

```python
# Plotting magnetiziation - T=1 L=20
plt.plot(n_cycles_20, magn_1_20, color="red",label="Initially random")
plt.plot(n_cycles_20_ord, magn_1_20_ord, color="blue",label="Initially ordered")

# add axis labels and legend
plt.xlabel("Number of MC cycles")
plt.ylabel(r"$\langle|$m$|\rangle$")
plt.legend(loc='lower right')

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)
plt.title("T=1 [J/$k_b$] L=20")
plt.xscale("log")

# save plot as pdf
plt.savefig("magn_1_20.pdf")
```

```python
# Plotting magnetiziation - T=2.4 L=20
plt.plot(n_cycles_20, magn_2_20, color="red",label="Initially random")
plt.plot(n_cycles_20_ord, magn_2_20_ord, color="blue",label="Initially ordered")

# add axis labels and legend
plt.xlabel("Number of MC cycles")
plt.ylabel(r"$\langle|$m$|\rangle$")
plt.legend(loc='upper right')

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)
plt.title("T=2.4 [J/$k_b$] L=20")
plt.xscale("log")
#plt.ylim(0.4,0.5)

# save plot as pdf
plt.savefig("magn_2_20.pdf")
```
