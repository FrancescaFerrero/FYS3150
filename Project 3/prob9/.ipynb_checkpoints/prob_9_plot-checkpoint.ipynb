```python
# multidimensional arrays
import numpy as np
import math

# inline plots
import matplotlib.pyplot as plt
%matplotlib inline

# nicer figures
import matplotlib as mpl
mpl.rcParams['axes.titlesize'] = 16
mpl.rcParams['axes.labelsize'] = 14
mpl.rcParams['xtick.labelsize'] = 12
mpl.rcParams['ytick.labelsize'] = 12
mpl.rcParams['legend.fontsize'] = 7.5
plt.rcParams["figure.figsize"] = (7,5)

#3d plot
from mpl_toolkits import mplot3d
```

```python
#reading time t
t = np.loadtxt("time.txt", usecols=0, dtype='double')

#reading r_z values
r_z = np.loadtxt("r1_nointer.txt", usecols=2, dtype='double')
```

```python
# Plotting motion in the z direction as a function of time
plt.plot(t, r_z)

# add axis labels
plt.xlabel("Time [$\mu s$]")
plt.ylabel("Motion in z direction")

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)

# save plot as pdf
plt.savefig("plot_9a.pdf")
```

```python
#confront between omega values
T = 100/11 # period

omega_teo = math.sqrt(2*9.65/40.08) # square root of 2qVo/md^2
omega_exp = 2*math.pi/T 

print("theoretical value",omega_teo)
print("experimental value",omega_exp)
```

```python
#with interaction
#reading r values two particles
r_x_1 = np.loadtxt("r1_inter.txt", usecols=0, dtype='double')
r_y_1 = np.loadtxt("r1_inter.txt", usecols=1, dtype='double')
r_z_1 = np.loadtxt("r1_inter.txt", usecols=2, dtype='double')

r_x_2 = np.loadtxt("r2_inter.txt", usecols=0, dtype='double')
r_y_2 = np.loadtxt("r2_inter.txt", usecols=1, dtype='double')
r_z_2 = np.loadtxt("r2_inter.txt", usecols=2, dtype='double')
```

```python
# Plotting motion of two particles with interactions
plt.plot(r_x_1, r_y_1, color="blue",label="particle 1")
plt.plot(r_x_2, r_y_2, color="red",label="particle 2")

# add axis labels and legend
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)
```

```python
#no interaction
#reading r values two particles
r_x_1_no = np.loadtxt("r1_nointer.txt", usecols=0, dtype='double')
r_y_1_no = np.loadtxt("r1_nointer.txt", usecols=1, dtype='double')
r_z_1_no = np.loadtxt("r1_nointer.txt", usecols=2, dtype='double')

r_x_2_no = np.loadtxt("r2_nointer.txt", usecols=0, dtype='double')
r_y_2_no = np.loadtxt("r2_nointer.txt", usecols=1, dtype='double')
r_z_2_no = np.loadtxt("r2_nointer.txt", usecols=2, dtype='double')
```

```python
diff=np.array(r_y_1-r_y_1_no)
zero = np.count_nonzero(diff)
print(zero)
```

```python
# Plotting motion of two particles with no interactions
plt.plot(r_x_1_no, r_y_1_no, color="blue",label="particle 1")
plt.plot(r_x_2_no, r_y_2_no, color="red",label="particle 2")

# add axis labels and legend
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)

# save plot as pdf
plt.savefig("plot_9b.pdf")
```

```python
#with interaction
#reading v values two particles
v_x_1 = np.loadtxt("v1_inter.txt", usecols=0, dtype='double')
v_y_1 = np.loadtxt("v1_inter.txt", usecols=1, dtype='double')
v_z_1 = np.loadtxt("v1_inter.txt", usecols=2, dtype='double')

v_x_2 = np.loadtxt("v2_inter.txt", usecols=0, dtype='double')
v_y_2 = np.loadtxt("v2_inter.txt", usecols=1, dtype='double')
v_z_2 = np.loadtxt("v2_inter.txt", usecols=2, dtype='double')
```

```python
#no interaction
#reading r values two particles
v_x_1_no = np.loadtxt("v1_nointer.txt", usecols=0, dtype='double')
v_y_1_no = np.loadtxt("v1_nointer.txt", usecols=1, dtype='double')
v_z_1_no = np.loadtxt("v1_nointer.txt", usecols=2, dtype='double')

v_x_2_no = np.loadtxt("v2_nointer.txt", usecols=0, dtype='double')
v_y_2_no = np.loadtxt("v2_nointer.txt", usecols=1, dtype='double')
v_z_2_no = np.loadtxt("v2_nointer.txt", usecols=2, dtype='double')
```

```python
# Plotting phase plot with no interactions
plt.plot(r_x_1_no, v_x_1_no, color="blue",label="particle 1")
plt.plot(r_x_2_no, v_x_2_no, color="red",label="particle 2")

# add axis labels and legend
plt.xlabel("x")
plt.ylabel("$v_x$")
plt.legend()

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)
```

```python
# Plotting phase plot with interactions
plt.plot(r_x_1, v_x_1, color="blue",label="particle 1")
plt.plot(r_x_2, v_x_2, color="red",label="particle 2")

# add axis labels and legend
plt.xlabel("x")
plt.ylabel("$v_x$")
plt.legend()

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)
```

```python
# 3D plot with interaction
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.scatter3D(r_x_1, r_y_1, r_z_1, s=1,color="blue",label="particle 1");
ax.scatter3D(r_x_2, r_y_2, r_z_2, s=1,color="red",label="particle 2");


ax.legend()
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
```

```python
# 3D plot with no interaction
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.scatter3D(r_x_1_no, r_y_1_no, r_z_1_no, s=1,color="blue",label="particle 1");
ax.scatter3D(r_x_2_no, r_y_2_no, r_z_2_no, s=1,color="red",label="particle 2");


ax.legend()
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
```
