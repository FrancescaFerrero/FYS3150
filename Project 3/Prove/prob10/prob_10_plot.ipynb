```python
# multidimensional arrays
import numpy as np
import math

# inline plots
import matplotlib.pyplot as plt
%matplotlib inline

# nicer figures
import matplotlib as mpl
mpl.rcParams['axes.titlesize'] = 16
mpl.rcParams['axes.labelsize'] = 14
mpl.rcParams['xtick.labelsize'] = 12
mpl.rcParams['ytick.labelsize'] = 12
mpl.rcParams['legend.fontsize'] = 7.5
plt.rcParams["figure.figsize"] = (7,5)

#3d plot
from mpl_toolkits import mplot3d
```

```python
#reading omega_v
omega = np.loadtxt("omega.txt", usecols=0, dtype='double')

#reading fraction of particles inside
n_f1 = np.loadtxt("p_inside_0.1.txt", usecols=0, dtype='double')
n_f2 = np.loadtxt("p_inside_0.4.txt", usecols=0, dtype='double')
n_f3 = np.loadtxt("p_inside_0.7.txt", usecols=0, dtype='double')
```

```python
# Plotting 
plt.plot(omega, n_f1, label="$f=0.1$")
plt.plot(omega, n_f2,label="$f=0.4$")
plt.plot(omega, n_f3,label="$f=0.7$")

# add axis labels
plt.xlabel("$\omega_v$")
plt.ylabel("Fraction of particle inside")
plt.legend()

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)

# save plot as pdf
plt.savefig("plot_10.pdf")
```
