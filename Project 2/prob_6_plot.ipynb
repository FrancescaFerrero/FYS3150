```python
# multidimensional arrays
import numpy as np
import math

#fitting
import scipy.stats as stats

# inline plots
import matplotlib.pyplot as plt
%matplotlib inline

# nicer figures
import matplotlib as mpl
mpl.rcParams['axes.titlesize'] = 16
mpl.rcParams['axes.labelsize'] = 14
mpl.rcParams['xtick.labelsize'] = 12
mpl.rcParams['ytick.labelsize'] = 12
mpl.rcParams['legend.fontsize'] = 12
plt.rcParams["figure.figsize"] = (7,5)
#mpl.rcParams['axes.spines.right'] = False
#mpl.rcParams['axes.spines.top'] = False
```

```python
N = np.loadtxt("iterations_6a.txt", usecols=0, dtype='double')
it = np.loadtxt("iterations_6a.txt", usecols=1, dtype='double')
```

```python
Nb = np.loadtxt("iterations_6b.txt", usecols=0, dtype='double')
itb = np.loadtxt("iterations_6b.txt", usecols=1, dtype='double')
```

```python
# plot results
plt.scatter(N,it,color="blue",s=15,label="num of iterations diagonal matrix")

x = np.linspace(10, 200, 200)
plt.plot(x,np.power(x,2),color="red",label="$N^2$")


# add axis labels and legend
plt.xlabel("N")
plt.ylabel("Iterations")
plt.legend()

# add a title 
plt.title("Number of iterations vs N")

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)

# save plot as pdf
plt.savefig("plot_6a.pdf")

```

```python
# plot results
plt.scatter(N,it,color="blue",s=18,label="num of iterations tridiagonal matrix")
plt.scatter(Nb,itb,color="orange",s=18,label="num of iterations dense matrix")
plt.plot(x,np.power(x,2),color="red",label="$N^2$")

# add axis labels and legend
plt.xlabel("N")
plt.ylabel("Iterations")
plt.legend()

# add a title 
plt.title("Number of iterations vs N")

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)

# save plot as pdf
plt.savefig("plot_6b.pdf")
```
