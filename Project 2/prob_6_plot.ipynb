```python
# multidimensional arrays
import numpy as np
import math

#fitting
import scipy.stats as stats

# inline plots
import matplotlib.pyplot as plt
%matplotlib inline

# nicer figures
import matplotlib as mpl
mpl.rcParams['axes.titlesize'] = 16
mpl.rcParams['axes.labelsize'] = 14
mpl.rcParams['xtick.labelsize'] = 12
mpl.rcParams['ytick.labelsize'] = 12
mpl.rcParams['legend.fontsize'] = 12
plt.rcParams["figure.figsize"] = (7,5)
#mpl.rcParams['axes.spines.right'] = False
#mpl.rcParams['axes.spines.top'] = False
```

```python
#reading file 
file = open('iterations.txt', 'r')
lines = file.readlines() 
file.close()

N=[]
it=[]

for line in lines:
    N.append(line.split()[0])
    it.append(line.split()[1])

for i in range(len(N)):
    N[i]=float(N[i])  
    
for i in range(len(it)):
    it[i]=float(it[i])
```

```python
#reading file 
file = open('iterations_dense.txt', 'r')
lines = file.readlines() 
file.close()

Nb=[]
itb=[]

for line in lines:
    Nb.append(line.split()[0])
    itb.append(line.split()[1])

for i in range(len(Nb)):
    Nb[i]=float(Nb[i])

for i in range(len(itb)):
    itb[i]=float(itb[i])
```

```python
# plot results
plt.scatter(N,it,color="blue",s=15,label="num of iterations diagonal matrix")

x = np.linspace(1, 200, 200)
plt.plot(x,np.power(x,2),color="red",label="$N^2$")


# add axis labels and legend
plt.xlabel("N")
plt.ylabel("Iterations")
plt.legend()

# add a title 
plt.title("Number of iterations vs N")

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)

# save plot as pdf
plt.savefig("plot_6a.pdf")

```
